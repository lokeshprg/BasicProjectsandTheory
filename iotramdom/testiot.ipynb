```python
# Day 27: Program 1

import cv2
import numpy as np

cap = cv2.VideoCapture(1)

while True:
    _, frame = cap.read()

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    lower_blue = np.array([60, 35, 140])
    upper_blue = np.array([180, 255, 255])

    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    result = cv2.bitwise_and(frame, frame, mask=mask)

    cv2.imshow("Original", frame)
    cv2.imshow("Mask", mask)
    cv2.imshow("Result", result)

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break
   
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        roi_gray = gray[y:y+h, x:x+w]

        
        smiles = smile_cascade.detectMultiScale(roi_gray,scaleFactor=1.8,minNeighbors=20,minSize=(25, 25))
        for (sx, sy, sw, sh) in smiles:
            cv2.rectangle(frame,(x + sx, y + sy),(x + sx + sw, y + sy + sh),(0, 255, 0),2)
    cv2.imshow("Smile Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# DAY 25:
import cv2
import datetime
vid = cv2.VideoCapture('vid1.mp4')
cv2.namedWindow("video", cv2.WINDOW_NORMAL)
cv2.setWindowProperty("video", cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)
while vid.isOpened():
    ret, frame = vid.read()
    if not ret:
        break
    font = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
    dt = str(datetime.datetime.now())
    frame = cv2.putText(frame, dt, (10, 100), font, 3, (255,0,0), 2, cv2.LINE_8)

    cv2.imshow('video', frame)
    key = cv2.waitKey(1)
    if key == ord('q') or key == 27:
        break
vid.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time

haar_cascade = 'cars.xml'
car_caascade = cv2.CascadeClassifier(haar_cascade)
if car_caascade.empty():
    raise Exception("Error loading haar cascade file.")

video = 'cars1.mp4'
cap = cv2.VideoCapture(video)

if not cap.isOpened():
    raise Exception("Error opening video file.")

car_detected = False
detection_time = None

cv2.namedWindow("video", cv2.WINDOW_NORMAL)
cv2.setWindowProperty("video", cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cars = car_caascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=8, minSize=(200,200))
        
    if len(cars)>0 and not car_detected:
        car_detected = True
        detection_time = time.time()
        print("Car detected!")

    if car_detected and time.time() - detection_time>10:
        break
        
    for (x, y, w, h) in cars:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,0,255), 2)
        cv2.imshow('video', frame)

    if cv2.waitKey(33) == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2 

cap = cv2.VideoCapture(0)
detector = cv2.QRCodeDetector()
print("Press 'q' to exit the live scanner loop.")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    data, bbox, _ = detector.detectAndDecode(frame)
    if bbox is not None:
        bbox = bbox.astype(int)
        n = len(bbox[0])
        for i in range(n):
            cv2.line(frame, tuple(bbox[0][i]), tuple(bbox[0][(i+1)% n]), (0,255,0), 2)

        if data:
            print(f"Decoded: {data}")
            cv2.putText(frame, data, (bbox[0][0][0], bbox[0][0][1] - 10), cv2.FONT_HERSHEY_COMPLEX, 1.5, (255, 0, 0), 2)

    cv2.imshow("Live QR Scanner", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
