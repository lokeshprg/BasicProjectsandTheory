```python
# Task 31
import cv2
import numpy as np
image = cv2.imread('download.png')
circle_image = cv2.circle(image, center=(100,100), radius=50, color=(0,255,0), thickness=12)

rect_image = cv2.rectangle(image, (5, 5), (220, 220), (255, 0, 0), 2)

text_img = cv2.putText(image, "LOKI", (40,40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)

line_image = cv2.line(image, (50,50), (250,50),(255,0,0),3)

pts1 = np.array([[50, 50], [150, 50], [100, 150]], np.int32).reshape((-1, 1, 2))
pts2 = np.array([[300, 300], [400, 300], [400, 400], [300, 400]], np.int32).reshape((-1, 1, 2))
poly_image = cv2.polylines(image, [pts1, pts2], isClosed=True, color=(0, 0, 255), thickness=2)

cv2.imshow('Circle', circle_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Task 32
import cv2
image = cv2.imread('photo.jpeg')
blurred_image = cv2.GaussianBlur(image, (5,5), 0)
rot_180 = cv2.rotate(image, cv2.ROTATE_180)
cv2.imshow("Avergin blur", blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Task 33
import cv2
import numpy as np
image = cv2.imread('photo.jpeg', cv2.IMREAD_GRAYSCALE)
blurred_image = cv2.GaussianBlur(image, (5,5), 0)
edges = cv2.Canny(blurred_image, threshold1=50, threshold2=150)
cv2.imshow('Canny Edges', edges)
cv2.waitKey()
cv2.destroyAllWindows()
```
