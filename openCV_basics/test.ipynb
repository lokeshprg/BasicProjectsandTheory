```python
# Visualizing different color channels of an image seperately
import cv2
img = cv2.imread("sample_parrot-image.jpg")

B, G, R = cv2.split(img)
cv2.imshow("Original",img)
cv2.imshow("Blue",B)
cv2.imshow("Green",G)
cv2.imshow("Red",R)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# Drawing a rectangle
import cv2

img = cv2.imread("roadTraffic.jpg")
copy = img.copy()
rectangle = cv2.rectangle(copy, (1500, 900), (600, 400), (0, 0, 255), 2)
cv2.imshow("Rectangle", rectangle)
cv2.imshow("Img", img)
cv2.imshow("Copy", copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Displaying Text
import cv2

img = cv2.imread("roadTraffic.jpg")
h, w = img.shape[:2]
ratio = 500/w
dim = (500, int(h*ratio))
resized_aspect = cv2.resize(img, dim)
copy = resized_aspect.copy()
text = cv2.putText(copy, "OpenCV demo", (100, 100), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 1)
cv2.imshow("Text", text)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Color Spaces in OpenCV
import cv2
img = cv2.imread("sample_parrot-image.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

cv2.imshow("Original Image", img)
cv2.imshow("RGB", img_rgb)
cv2.imshow("HSV", img_hsv)
cv2.imshow("GRAY", img_gray)
cv2.imshow("LAB", img_lab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Detecting red color in an image
import cv2
import numpy as np

img = cv2.imread("sample_parrot-image.jpg")
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])

mask = cv2.inRange(img_hsv, lower_red, upper_red)
result = cv2.bitwise_and(img, img, mask=mask)

cv2.imshow("Red", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Extracting region of interest (ROI)
import cv2

img = cv2.imread("roadTraffic.jpg")
roi = img[100:500, 200:700]
cv2.imshow("Original", img)
cv2.imshow("ROI", roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resizing the image
import cv2

img = cv2.imread("roadTraffic.jpg")
resize = cv2.resize(img, (100, 500))
cv2.imshow("Original Image", img)
cv2.imshow("Resized Image", resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# To maintain the aspect ratio of the image
import cv2

img = cv2.imread("roadTraffic.jpg")
w, h = img.shape[:2]
ratio = 100/w
dim = (100, int(h*ratio))
resize_aspect = cv2.resize(img, dim)
cv2.imshow("Resized", resize_aspect)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
